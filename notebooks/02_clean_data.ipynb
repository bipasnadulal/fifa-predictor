```python
import pandas as pd
import numpy as np
```

```python
matches_df = pd.read_csv('../data/results.csv')
rankings_df = pd.read_csv('../data/fifa_ranking.csv')

print("Matches: ", matches_df.shape)
print("Rankings: ", rankings_df.shape)

print("Matches Columns: ", matches_df.columns.tolist())
print("Rankings Columns: ", rankings_df.columns.tolist())
```

#### Checking missing values

```python
print("Missing values in Matches: \n", matches_df.isnull().sum())
print("Missing values in Rankings: \n", rankings_df.isnull().sum())

print(" ")
print("Date range of matches: ")
print("Earliest: ", matches_df['date'].min())
print("Latest: ", matches_df['date'].max())
```

##### Since the data ranges from 1872-2026, we only want the recent history to as matches from 1872 are not very useful, football has evolved since then.

```python
#proper date format
matches_df['date'] = pd.to_datetime(matches_df['date'])
print(matches_df['date'].head(3))
```

```python
# keeping only matches from 2014 onwards
matches_df = matches_df[matches_df['date'] >= '2014-01-01']

print("Matches after 2014:")
print(matches_df.shape)
print(matches_df.head())
```

```python
# Seeing tournament types
print(matches_df['tournament'].value_counts())

# adding a feature for friendly matches
matches_df['is_friendly'] = matches_df['tournament'] == 'Friendly'

print(" ")
print(matches_df['is_friendly'].value_counts())
```

### Getting the World cup 2026 teams

```python
world_cup_2026_teams = [

    # Hosts (CONCACAF)
    'Canada',
    'Mexico',
    'United States',

    # AFC (Asia)
    'Australia',
    'Iraq',
    'Iran',
    'Japan',
    'Jordan',
    'South Korea',
    'Qatar',
    'Saudi Arabia',
    'Uzbekistan',

    # CAF (Africa)
    'Algeria',
    'Cabo Verde',
    'Democratic Republic of Congo',
    "Côte d'Ivoire",
    'Egypt',
    'Ghana',
    'Morocco',
    'Senegal',
    'South Africa',
    'Tunisia',

    # CONCACAF
    'Curaçao',
    'Haiti',
    'Panama',

    # CONMEBOL (South America)
    'Argentina',
    'Brazil',
    'Colombia',
    'Ecuador',
    'Paraguay',
    'Uruguay',

    # OFC
    'New Zealand',

    # UEFA (Europe)
    'Austria',
    'Belgium',
    'Bosnia and Herzegovina',
    'Croatia',
    'Czechia',
    'England',
    'France',
    'Germany',
    'Netherlands',
    'Norway',
    'Portugal',
    'Scotland',
    'Spain',
    'Sweden',
    'Switzerland',
    'Türkiye'
]

print(f"Total teams: {len(world_cup_2026_teams)}")
```

#### Filtering matches to only World Cup teams

```python
wc_matches = matches_df[
    matches_df['home_team'].isin(world_cup_2026_teams) &
    matches_df['away_team'].isin(world_cup_2026_teams)
]

print(wc_matches.shape)
print(wc_matches.head())
```

```python
wc_matches.to_csv('../data/wc_matches.csv', index=False)
```
